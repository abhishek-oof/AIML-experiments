```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score , precision_score,recall_score ,f1_score,roc_auc_score,confusion_matrix , classification_report
import pandas as pd
```

```python
data = load_breast_cancer()
x = data.data
y = data.target

print("Classes:",data.target_names)

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3)
```

```python
lr = LogisticRegression(max_iter=5000)
lr.fit(x_train,y_train)
pred_lr = lr.predict(x_test)
print("Logistic Test Acurracy:",accuracy_score(y_test,pred_lr))
pred_lr_train = lr.predict(x_train)
print("Logistic Train Accuracy:",accuracy_score(y_train,pred_lr_train))
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
pred_dt = dt.predict(x_test)
print("Decision Tree Test Acurracy:",accuracy_score(y_test,pred_dt))
pred_dt_train = dt.predict(x_train)
print("Decicsion Train Accuracy:",accuracy_score(y_train,pred_dt_train))
```

```python
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
pred_knn = knn.predict(x_test)
print("KNN Tree Test Acurracy:",accuracy_score(y_test,pred_knn))
pred_knn_train = knn.predict(x_train)
print("KNN Train Accuracy:",accuracy_score(y_train,pred_knn_train))
```

```python
gs = GaussianNB()
gs.fit(x_train,y_train)
pred_gs = gs.predict(x_test)
print("GaussianNB Test Acurracy:",accuracy_score(y_test,pred_gs))
pred_gs_train = gs.predict(x_train)
print("GaussianNB Train Accuracy:",accuracy_score(y_train,pred_gs_train))
```

```python
nn = MLPClassifier()
nn.fit(x_train,y_train)
pred_nn = nn.predict(x_test)
print("Neural Network Tree Test Acurracy:",accuracy_score(y_test,pred_nn))
pred_nn_train = nn.predict(x_train)
print("Neural Network Train Accuracy:",accuracy_score(y_train,pred_nn_train))
```

```python
precision_lr = precision_score(y_test,pred_lr)
recall_lr = recall_score(y_test,pred_lr)
f1_lr = f1_score(y_test,pred_lr)
roc_lr = roc_auc_score(y_test,pred_lr)

print("Logistic Regression:")
print("Precision:",precision_lr)
print("Recall:",recall_lr)
print("f1:",f1_lr)
print("Roc:",roc_lr)
```
